# 3 数据处理

```python
# 创建空列表
title_list = []
```

```python
# 换了一种读取方法
with open('D:/input/data.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()
    for line in lines:
        line = line.strip()
        product_items = line.split(',')
        for item in product_items:
            title_list.append(item)
```

```python
# 使用set数据结构去重
title_list = set(title_list)
```

```python
# 打印商品列表
title_list
```

```python
# 打印商品列表长度
print("共有", len(title_list) ,"种商品")
```

```python
# 关闭文件
origin_data.close()
```

```python
# 创建一个空的dataframe用来存放处理后的数据集
title_list = list(title_list)
changed_data = pd.DataFrame(columns=title_list)
# 此处原本有一个ValueError: columns cannot be a set报错，应当按上面方式修改，把set(unordered)转化为list

# 打开文件
origin_data = open("D:/input/data.txt", "r", encoding='utf-8')
```

```python
import csv

# Create an empty dictionary to store the customer records
customers = []

# Open the file using the csv module
with open("D:/input/data.txt", "r", encoding="utf-8") as file:
    # Create a csv reader object
    reader = csv.reader(file)

    # Iterate over each row in the file
    for row in reader:
        # Remove leading/trailing whitespaces from each item in the row
        row = [item.strip() for item in row]

        # Create a dictionary to store the customer's purchase information
        customer = {item: 1 if item in row else 0 for item in title_list}

        # Append the customer record to the list
        customers.append(customer)

# Create a DataFrame from the list of customer records
changed_data = pd.DataFrame(customers)
```

```python
# 将处理后的数据集保存
changed_data.to_excel("D:/output/changed_data.xlsx", index=False)
```

# 关联规则挖掘

```python
changed_data = changed_data.astype(bool)
# 解决报错 DataFrames with non-bool types result in worse computationalperformance and their support might be discontinued in the future.Please use a DataFrame with bool type
frequent_itemsets = apriori(changed_data, min_support=0.05, use_colnames=True)
```

```python
# 输出频繁项集
frequent_itemsets
```

```python
rules_1 = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_1
```

```python
frequent_itemsets
```

```python
rules_2 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules_2
```

```python
rules_3 = association_rules(frequent_itemsets, metric="support", min_threshold=0.05)
rules_3
```
